```
# IMAGE RESIZER

from PIL import Image
import os, sys

opath = base_dir+'thankas/original/'
dirs = os.listdir( opath )


def resize(dst, x=256, y=256):
    path = opath
    dpath = base_dir + 'thankas/' + dst
    for item in dirs:
        if os.path.isfile(opath+item):
            im = Image.open(opath+item)
            f, e = os.path.splitext(opath+item)
            imResize = im.resize((x,y), Image.ANTIALIAS)
            print ("HERE", item)
            imResize.save(dpath + item, 'JPEG', quality=90)

```
